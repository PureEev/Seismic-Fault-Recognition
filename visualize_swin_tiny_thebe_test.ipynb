```python
import plotly.io as pio
import numpy as np
import plotly.graph_objects as go
```

```python
gt = np.load("/home/burlaka-na/Desktop/vuz/diplom/gt.npy")
preds = np.load("/home/burlaka-na/Desktop/vuz/diplom/preds.npy")
```

```python
def visualize_3d_bool(predicted, gt, outfile="3d_compare.html", max_points=200_000):
    assert predicted.shape == gt.shape

    tp = predicted & gt
    fp = predicted & (~gt)
    fn = (~predicted) & gt

    def points(mask, max_points=max_points):
        pts = np.argwhere(mask)
        if len(pts) > max_points:
            pts = pts[np.random.choice(len(pts), max_points, replace=False)]
        return pts.T if len(pts) else ([], [], [])

    traces = []

    x, y, z = points(tp)
    traces.append(
        go.Scatter3d(
            x=x, y=y, z=z,
            mode="markers",
            marker=dict(size=1, color="green", opacity=1.0),
            name="True Positive (match)"
        )
    )

    x, y, z = points(fp)
    traces.append(
        go.Scatter3d(
            x=x, y=y, z=z,
            mode="markers",
            marker=dict(size=1, color="red", opacity=0.1),
            name="False Positive (pred only)"
        )
    )

    x, y, z = points(fn)
    traces.append(
        go.Scatter3d(
            x=x, y=y, z=z,
            mode="markers",
            marker=dict(size=1, color="blue", opacity=0.1),
            name="False Negative (gt only)"
        )
    )

    fig = go.Figure(traces)
    fig.update_layout(
        title="3D comparison: predicted vs gt",
        scene=dict(aspectmode="data")
    )

    fig.write_html(outfile)
    print(f"Saved to {outfile}")

visualize_3d_bool(preds[:, 800:, 750:], gt[:, 800:, 750:])
```
